### Load Data

```python
# Import necessary libaries
import pandas as pd 

# Import Movie Metadata dataset 
Movie_MetaData_Reviews= pd.read_csv('Merged_Dataset.csv')

# Show the top 10 records of the DataFrame 
Movie_MetaData_Reviews.head(10)
```

```python
# Read training dataset into a DataFrame
movieReviewTraining = pd.read_csv('archive/MovieReviewTrainingDatabase.csv')

# Display the first 10 rows of the DataFrame
movieReviewTraining.head(10)
```

### Data cleaning / cleansing (removal of unwanted variables)

```python
# Split the release date column into two new columns
Movie_MetaData_Reviews[['release_date', 'release_country']] = Movie_MetaData_Reviews['release_date'].str.split(' \(', expand=True)

# Remove the parenthesis from the 'release_country' column
Movie_MetaData_Reviews['release_country'] = Movie_MetaData_Reviews['release_country'].str.replace('[()]', '', regex=True)
# Show data column head
Movie_MetaData_Reviews.head(10)
```

```python
# Import necessary libraries 
import numpy as np
# Check for nulls and replace with NaN to catch any missing data
Movie_MetaData_Reviews = Movie_MetaData_Reviews.replace('Null', np.nan)
# Print a sum of all null values to find missing values
Movie_MetaData_Reviews.isnull().sum()
```

```python
# Check for nulls and replace with NaN to catch any missing data 
movieReviewTraining = movieReviewTraining.replace('Null', np.nan)
# Print a sum of all null values to find missing values
movieReviewTraining.isnull().sum()
```

```python
# Drop any rows from the Movie_MetaData_Reviews DataFrame that contain missing values
Clean_Movie_MetaData_Reviews = Movie_MetaData_Reviews.dropna()

# Print the sum of all null values in the Clean_Movie_MetaData_Reviews DataFrame
# This will help identify if there are any remaining missing values in the cleaned dataset
Clean_Movie_MetaData_Reviews.isnull().sum()
```

```python
# Filter the columns to keep only the desired ones
Clean_Movie_MetaData_Reviews = Clean_Movie_MetaData_Reviews[['name', 'movie_rated', 'run_length', 'Combined Genre',
                                                 'release_date', 'rating_x', 'rating_y', 'date', 'review', 'release_country']]

# Rename ambiguous columns
Clean_Movie_MetaData_Reviews = Clean_Movie_MetaData_Reviews.rename(columns={'rating_x': 'IMDB_rating',
                                                                'rating_y': 'reviewer_rating',
                                                                'date': 'review_date'})
# Show the first five rows of the new data frame 
Clean_Movie_MetaData_Reviews.head()
```

```python
# Get the shape of the Clean_Movie_MetaData_Reviews DataFrame
Clean_Movie_MetaData_Reviews.shape
```

```python
# Get the shape of the movieReviewTraining DataFrame
movieReviewTraining.shape
```

```python
# Print detailed information about the Clean_Movie_MetaData_Reviews DataFrame
Clean_Movie_MetaData_Reviews.info()
```

```python
# Change data types for statistics

# Change reviewer_rating and IMDB_rating to an int
Clean_Movie_MetaData_Reviews['reviewer_rating'] = pd.to_numeric(Clean_Movie_MetaData_Reviews['reviewer_rating'])

# Change release_date and review_date to a date and time
Clean_Movie_MetaData_Reviews['release_date'] = pd.to_datetime(Clean_Movie_MetaData_Reviews['release_date'], format='%d %B %Y')
Clean_Movie_MetaData_Reviews['review_date'] = pd.to_datetime(Clean_Movie_MetaData_Reviews['review_date'], format='%d %B %Y')

# Validate new datatypes
Clean_Movie_MetaData_Reviews.info()
```

```python
# Write clean dataset to CSV to be used for the rest of the project 
Clean_Movie_MetaData_Reviews.to_csv("clean_movie_metadata_reviews.csv", index=False)
```

### Data visualization 

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Create a figure and axes for subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot the IMDB rating histogram
sns.histplot(data=Clean_Movie_MetaData_Reviews, x='IMDB_rating', bins=20, kde=True, color='blue', ax=ax1)
ax1.set_xlabel('IMDB Rating')
ax1.set_ylabel('Count')
ax1.set_title('Distribution of IMDB Ratings')

# Plot the reviewer rating histogram
sns.histplot(data=Clean_Movie_MetaData_Reviews, x='reviewer_rating', bins=20, kde=True, color='red', ax=ax2)
ax2.set_xlabel('Reviewer Rating')
ax2.set_ylabel('Count')
ax2.set_title('Distribution of Reviewer Ratings')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Extract the year from the datetime column
Clean_Movie_MetaData_Reviews['Year'] = Clean_Movie_MetaData_Reviews['release_date'].dt.year

# Count the number of movies for each year and release country
movies_per_year_country = Clean_Movie_MetaData_Reviews.groupby(['Year', 'release_country'])['name'].count().reset_index()
movies_per_year_country.columns = ['Year', 'Country', 'Count']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(16, 8))

# Plot the number of movies released by year and country
sns.lineplot(data=movies_per_year_country, x='Year', y='Count', hue='Country', ax=ax)

# Set plot title and labels
ax.set_title('Movies Released by Year and Country', fontsize=16, fontweight='bold')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Number of Movies', fontsize=12)

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Adjust the plot layout and display it
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Count the number of movies for each genre
movies_per_genre = Clean_Movie_MetaData_Reviews['Combined Genre'].value_counts().sort_index()

# Create a DataFrame with the genre and count
genre_data = pd.DataFrame({'Genre': movies_per_genre.index, 'Count': movies_per_genre.values})

# Calculate the average IMDB rating and user rating for each genre
imdb_rating_per_genre = Clean_Movie_MetaData_Reviews.groupby('Combined Genre')['IMDB_rating'].mean().reset_index()
imdb_rating_per_genre.columns = ['Genre', 'IMDB_rating']

user_rating_per_genre = Clean_Movie_MetaData_Reviews.groupby('Combined Genre')['reviewer_rating'].mean().reset_index()
user_rating_per_genre.columns = ['Genre', 'User_rating']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(18, 8))

# Plot the number of movies for each genre
ax2 = ax.twinx()  # Create a second y-axis
ax2.set_ylabel('Average Rating', fontsize=12)
sns.barplot(data=genre_data, x='Genre', y='Count', ax=ax)
ax.set_ylabel('Number of Movies', fontsize=12)
ax.set_title('Movies by Genre with Average Ratings', fontsize=16, fontweight='bold')

# Plot the average IMDB rating for each genre
sns.lineplot(data=imdb_rating_per_genre, x='Genre', y='IMDB_rating', linewidth=2.5, ax=ax2, color='blue', label='IMDB Rating')

# Plot the average user rating for each genre
sns.lineplot(data=user_rating_per_genre, x='Genre', y='User_rating', linewidth=2.5, ax=ax2, color='red', label='User Rating')

# Add legend
ax2.legend(loc='upper left', fontsize=12)

# Rotate x-axis labels if needed
plt.xticks(rotation=90)

# Remove top and right spines
sns.despine(top=True,right=False)

# Display the plot
plt.show()
```
